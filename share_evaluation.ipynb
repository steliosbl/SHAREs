# Survival SHAREs Evaluation

We evaluate the top-scoring models from the survival SHAREs experiment.

```python
import numpy as np 
import pandas as pd 
import torch 
import matplotlib.pyplot as plt
import seaborn as sns

import os 
from pathlib import Path
from datetime import datetime

sns.set_style("whitegrid")
sns.set_palette("colorblind")
torch.cuda.is_available()
```

## 0. Datasets

```python
from datasets import Rossi, Metabric, GBSG2
from sklearn.model_selection import train_test_split

##### Data #####
dataset = dict(
    rossi=Rossi(), # Cox C-index 0.6403292470997135
    metabric=Metabric(), # Cox C-index 0.6391503770136181
    gbsg2=GBSG2(), # Cox C-index 0.5437638695233794
)['rossi']

X, T, E = dataset.load(normalise=False)
feature_names = dataset.features

X_train, X_test, T_train, T_test, E_train, E_test = train_test_split(
    X, T, E, test_size=0.2, random_state=42
)
```

## 1. Rossi Data

```python
variant = 'rossi_pll_shrink'
checkpoint_dir = Path(f'data/checkpoints/{variant}')
timestamp = max(os.listdir(checkpoint_dir), key=lambda x: datetime.strptime(x, "%Y-%m-%dT%H.%M.%S"))
```

```python
res_df = pd.read_csv(checkpoint_dir / timestamp / "output.csv")
res_df = res_df[res_df.n_shapes != 0].sort_values(by="brier_test", ascending=True).reset_index(drop=True)
res_df.head(10)
```

```python
from experiments.utils import create_df_from_cached_results, load_share_from_checkpoint

eq = res_df.iloc[0]['equation']
id = res_df.iloc[0]['id']

esr_best = load_share_from_checkpoint(
    timestamp, eq, 
    checkpoint_dir=checkpoint_dir, 
    task='regression',
    n_features=len(feature_names), 
    equation_id=id
)
esr_best._program.feature_names = feature_names
esr_best
```

```python
from sklearn.base import BaseEstimator, RegressorMixin
from pycox.models.cox import _CoxPHBase
import torchtuples as tt
class SymRegPH(BaseEstimator, RegressorMixin, _CoxPHBase):
    """
    Wrapper for gplearn's SymbolicRegressor to use with pycox supporting functions
    """

    def __init__(self, model):
        self.model = model

    def predict(self, X, *args, **kwargs):
        if isinstance(X, tt.TupleTree):
            X = X[0]
        return self.model.predict(X)
```

```python
esr_wrapped = SymRegPH(esr_best)
esr_wrapped.compute_baseline_hazards(X_train, (T_train, E_train))
surv_pred = esr_wrapped.predict_surv_df(X_test)
```

```python
from plot import survival_probability_calibration

survival_probability_calibration(surv_pred, T_test, E_test, t0=Rossi().tmax)
```
